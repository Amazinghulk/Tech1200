```python
while True:
  print("n/simple calculator")
  print("enter 'x' to exit")
  num1=input("enter the first number:")
  if num1=="x":
    print("calculator terminated")
    break
  num1=float(num1)
  operator=input("enter operator(+,-,*,%):")
  num2=input("enter the second number:")
  if num2=="x":
    print("calculator terminated")
    break
  num2=float(num2)
  if operator=='+':
   result=num1+num2
   print("result:",result)
  elif operator=='-':
    result=num1-num2
    print("result:",result)
  elif operator=='*':
    result=num1*num2
    print("result:",result)
  elif operator=='%':
    result=num1%num2
    print("result:",result)
  else:
    print("invalid operator")
    print("result:",result)

```

```python
# Initialize sum
total = 0

# Prompt user for a number
number = int(input("Enter a number: "))

# Repeat until user enters 0
while number != 0:
    total += number
    number = int(input("Enter a number: "))

# Display the total sum
print("Total sum is:", total)
11
```
