```python
#Assignemnt-1[Student Management System by using Python]#
students=[]#blank list
grades=[] # grade list

def add_student(students, grades):#define a new student
   name=input("enter student name:")#name from user input
   age=int(input("enter student age"))
#convert input tp float to handle decimal grades
   grade=float(input("enter student marks"))
   if grade<0 or grade>100:
    print("invalid marks,marks should be between 0 and 100")
    return#exit

   student=[name,age,grade]
   students.append(student)#append to add the student to list
   grades.append(grade)#adding grade to list
   print("name added perfectly. with grade!")

def view_student(students,grades):
  print("-----student record---")
  if not students:
      print("No students to display.")
      return
  for a, student_cell in enumerate(students, start=1):#to iterate over a sequence
      print(f"{a}. name:{student_cell[0]}, age:{student_cell[1]}, grade:{student_cell[2]}")
#f-string is used formatted string literal in python
def update_student(students, grades):
    print("-----update student record---")
    name=input("enter student name for update:")
    for a, student_cell in (students):
        if student_cell[0] == name:
          new_grade=float(input("enter new grade:"))
          if new_grade<0 or new_grade>100:
            print("invalid grade. should not be between 0 and 100")
            return
          student_cell[2] = new_grade # Update grade in student record
          grades[a] = new_grade # Update grade in grades list
          print(f"Student {name}'s grade updated to {new_grade}.")
          break
    else:
        print(f"Student {name} not found.")



def remove_student(students, grades):
   print("-----remove student-----")
   name=input("enter student name to remove:")
   for a, student_cell in(students):
     if student_cell[0] == name:
       del students[a]#to delete student
       del grades[a]#to delete grade
       print(f"student{name} removed successfully")
       break
     else:#to remove student successfully if not student name is not in the derived list
       print(f"Student{name} not found.")

def get_grade_cell(grade):
    #categorise into A,B,C,D or F based on given ranges
    #A=super
    #B=ok
    #C=average
    #D=can do better
    #f=failure is the stepping stone of success
    if grade>=85:
      return "A"
    elif grade>=75:
      return "B"
    elif grade>=50:
      return "C"
    elif grade>=35:
      return"D"
    else:
      return "F"

def categorize_student(students, grades):
    print("------grade categorise------")
    if not students:
      print("No students to categorize.")
      return

    categories = {"A":[], "B":[], "C":[], "D":[], "F":[]}
    for student_cell, in students:
        name = student_cell[0]
        grade = student_cell[2]
        category = get_grade_cell(grade)
        categories[category].append(name)

    for category, student_list in categories.items():#some variables are taken
       if student_list:
        print(f"\n{category} grades:")
        for student_name in student_list:
          print(f"\n{student_name}")
        if category=="A":
          print("  Super mate!")
        elif category=="B":
          print("  Ok!")
        elif category=="C":
          print("  Average!")
        elif category=="D":
          print("  Can do better!")
        elif category=="F":
          print("  Failure is the stepping stone of success!")
       else:
          print(f"\n{category} No students in this category.")

def display_menu():
    print("\n--- Student Management System ---")
    print("1. Add student")
    print("2. View students")
    print("3. Update student grade")
    print("4. Remove student")
    print("5. Categorize students by grade")
    print("6. Exit")

#main program
def main():
   students=[] #list to store student names
   grades=[]  #list to store student grades

   while True:
    display_menu()
    option=input("Enter your option:")

    if option=="1":
      add_student(students, grades)

    elif option=="2":
      view_student(students, grades)

    elif option=="3":
      update_student(students, grades)

    elif option=="4":
      remove_student(students, grades)

    elif option=="5":
      categorize_student(students, grades)
    elif option=="6":
      print("thank you for using!")
      print("done!")
      break
    else:
      print("Invalid option. Please try again later.")

if __name__=="__main__":
  main()
```
